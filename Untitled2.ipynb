```python
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import MinMaxScaler

pred_transformer = RobustScaler()
pred_transformer = pred_transformer.fit(train[['1']])
train['1'] = pred_transformer.transform(train[['1']])
test['1'] = pred_transformer.transform(test[['1']])
# scaler = MinMaxScaler(feature_range=(0, 1))
# scaler.fit(train[['1']])
# train = scaler.fit_transform(train['1'])
# test =  scaler.fit_transform(test['1'])
```

```python
def create_dataset(X, y, time_steps=1):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        v = X.iloc[i:(i + time_steps)].values
        Xs.append(v)
        ys.append(y.iloc[i + time_steps])
    return np.array(Xs), np.array(ys)
```

```python
TIME_STEPS = 30
X_train, y_train = create_dataset(train, train['1'], time_steps=TIME_STEPS)
X_test, y_test = create_dataset(test, test['1'], time_steps=TIME_STEPS)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
X_train[0][0].shape
```

```python
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from tensorflow.keras import preprocessing
from tensorflow.keras.layers import Bidirectional, Dropout, Activation, Dense, LSTM
from tensorflow.python.keras.layers import CuDNNLSTM
from tensorflow.keras.models import Sequential
```

```python
 
model = keras.Sequential()
model.add(
  # keras.layers.Bidirectional(
  #   keras.layers.LSTM(
  #     units=100,
  #     input_shape=(X_train.shape[1], X_train.shape[2])
  #   )
  # )
  LSTM(100,return_sequences=True,input_shape=(X_train.shape[1], X_train.shape[2]))
)

model.add(LSTM(100,return_sequences=True))
model.add(LSTM(100))
model.add(keras.layers.Dropout(rate=0.2))
model.add(keras.layers.Dense(units=1))
model.compile(loss='mean_squared_error', optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(
    X_train, y_train,
    epochs = 10,
    batch_size=64,
    validation_split=0.1,
    shuffle = False
)
```

```python
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='validation')
plt.legend()
```

```python
y_pred_test = model.predict(X_test)
y_pred_train = model.predict(X_train)
```

```python
y_train_inv = pred_transformer.inverse_transform(y_train.reshape(1,-1))
y_test_inv = pred_transformer.inverse_transform(y_test.reshape(1,-1))
y_pred_test_inv = pred_transformer.inverse_transform(y_pred_test)
y_pred_train_inv = pred_transformer.inverse_transform(y_pred_train)
```

```python
# Calculating MSE
print(y_train_inv[:10])
print(y_test_inv[:10])
print(y_pred_test_inv[:10])
print(y_pred_train_inv[:10])
```

```python
from sklearn.metrics import mean_squared_error
import math
print(y_train_inv[0])
print( y_pred_train_inv[:,0])
trainScore = math.sqrt(mean_squared_error(y_train_inv[:,0], y_pred_train_inv[0]))
print('Train Score: %.2f RMSE' % (trainScore))
testScore = math.sqrt(mean_squared_error(y_test_inv[:,0], y_pred_train_inv[0]))
print('Test Score: %.2f RMSE' % (testScore))
```
